Exploring conservation categories that are not "no intervention" through stacked bar chart


```python
conservationCategory = species[species.conservation_status != "No Intervention"].groupby(['conservation_status', "category"])["scientific_name"].count().unstack()
conservationCategory
```

```python
ax = conservationCategory.plot(kind = "bar", figsize = (8,6), stacked = True)
ax.set_xlabel("Conservation Status")
ax.set_ylabel("Number of Species")
plt.title("Conservation Status Count")
```

What types of species are most likely to be endangered/require conservation?

```python
species["endangered"] = species.conservation_status != "No Intervention"
species.head()
```

```python
protection_status = species.groupby(["category", 'endangered'])\
    .scientific_name.nunique()\
    .reset_index()\
    .pivot(columns = 'endangered',
          index = 'category',
          values = 'scientific_name')\
    .reset_index()
protection_status.columns = ["category", "not_protected", "protected"]

protection_status
```

```python
protection_status["percent_protected"] = (protection_status.protected/(protection_status.protected + protection_status.not_protected))*100
protection_status.head(10)
```

# Statistical Significance


```python
from scipy.stats import chi2_contingency

contingency1 = [[30, 146], 
               [75, 413]]

chi2_contingency(contingency1)
```

From the results of the chi-squared test above, we observe a p-value of 0.69. Given that the standard p-value to test statistical significance by is 0.5, and the resulting p-value (0.69) is greater than 0.5m we can say that there does not seem to be any significant relationship between mammals and birds (both variables are independent from each other)

```python
contingency_2 = [[30, 146], 
                [5, 73]]

chi2_contingency(contingency_2)
```

Looking at Species in Parks

Givem that the column "common_names" in the species dataframe contains a string of string names for each species, we will need to remove the punctuation separating each name, and convert it into a list 


```python
from itertools import chain
import string

#lets achieve this by removing the punctuation with a function:
def remove_punctuations(text):
    for punctuation in string.punctuation: 
        text = text.replace(punctuation, "")
    return text
common_names = species[species.category == "Mammal"]\
    .common_names\
    .apply(remove_punctuations)\
    .str.split().tolist()

common_names[:6]
```

We remove any duplicate words in each row

```python
clean_rows = []
from collections import OrderedDict

#we use dict.fromkeys as dictionaries can not have multiple keys with the same value, 

for item in common_names:
    row = list(dict.fromkeys(item))
    clean_rows.append(row)
clean_rows[:6]
```

We will now combine all lists into one list for easier use:

Note that these animals are particularly for the mammal species. 

```python
mammal_data = list(chain.from_iterable(i if isinstance(i, list) else [i] for i in clean_rows))
mammal_data
```

We will count the number of occurences of each animal/species using the mammal_data above

```python
species_counted = []

for animal in mammal_data: 
    animal_count = mammal_data.count(animal)
    species_counted.append((animal, animal_count))

pd.DataFrame(set(species_counted), columns = ["Word", "Count"]).sort_values("Count", ascending = False).head(10)
```

```python
# We will be creating an is_bat column to see which rows of species are referring to bats

species["is_bat"] = species.common_names.str.contains(r"\bBat\b", regex = True)
species.head(50)
```

# Merging Observations Data and species.is_bat Data

```python
bat_observations = observations.merge(species[species.is_bat])
#note merge will drop all instances in which bat is not True; therefore this new dataframe is ALL ABOUT BATS!!!
bat_observations.head(150)
```

```python
bat_observations.groupby("park_name").observations.sum()
```

```python
observation_by_park = bat_observations.groupby(["park_name", "endangered"]).observations.sum().reset_index()

print(observation_by_park)

#for easy readibility
# observation_by_park_pivot = observation_by_park.pivot(
#     index = "park_name", 
#     columns = "endangered",
#     values = "observations").reset_index()
# observation_by_park_pivot 
```

# Plotting Bat observation per week

```python
plt.figure(figsize = (16,4))
sns.barplot(x = observation_by_park.park_name, y = observation_by_park.observations, hue=observation_by_park.endangered)
plt.xlabel("National Parks")
plt.ylabel("Number of Observations")
plt.show()
```
